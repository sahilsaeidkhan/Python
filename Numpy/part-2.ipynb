# Sorting

```python
#basic sorting

arr = np.array([9,8,6,34,2,5,34,56,12])
print("Sorted array:",np.sort(arr))


arr_2d = np.array([[3,4,5],[6,4,3],[5,67,3]])
#to sort rows 
print("Sorted rows",np.sort(arr_2d,axis=0))

#to sort column
print("Sorted column",np.sort(arr_2d,axis=1))
```

# Filter

```python
num = np.array([3,4,5,6,7,8,9])

print("Odd numbers",num[num%2 != 0])

# we can put whole expression inside []
```

## Filter with mask


```python
mask = num<6
print(num[mask])

# we can store whole expression inside variable and use in code 
```

# Indices


```python
numbers = [0,2,4] # stored the indices

print(num[numbers]) # used the indices to find numbers


where_result = np.where(num>6) # get the index of the numbers which are > 6

print("numbers are" , num[where_result]) #used the above index to find numbers which are >6

```

```python
# where existing array ko modify krta hai 

condition_array = np.where(num>5 , num*2  , num) # if num is greater than 5 , multiply it with 2 else print as it is num 
# it works like if num>5 print(mutliply by 2) else num
 
print(condition_array)

```

# Adding and concatenating arrays


```python
first = np.array([3,4,5,6])
second = np.array([5,3,6,3])

print(first + second) # here values are added example 3 will be added to 5 = 8 

combined = np.concatenate((first,second))
print(combined) # here arrays are concatenated 
```
