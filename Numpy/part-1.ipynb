Array Reshaping

```python
arr = np.arange(10)

print("Reshaped array:" , arr.reshape((2,5)))

print("Flattened array:",arr.flatten())

# returns original array and above flatten , reshape return copy of the array 
print("Ravelled array:",arr.ravel())

print("Transposed array:",arr.T)


```
