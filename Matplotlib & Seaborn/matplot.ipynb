```python
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.plot(x,y)
plt.title("Sales growth")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.plot(x,y,color='red', linestyle='--', marker='o')
plt.show()
```

```python
categories = ["A","B","C"]
values = [10,20,15]

plt.bar(categories,values)
plt.show()
```

```python
import numpy as np

data = np.random.randn(100)

plt.hist(data)
plt.show()
```

```python
sizes = [30,40,30]
labels = ["Python","Java","C++"]

plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.show()
```
